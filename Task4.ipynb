<h2 align='center'> FM </h2>

```python
import matplotlib.pylab as plt
import numpy as num
import plotly.express as px
```

```python
fm=5
fc=25
modulation_index=10
time=num.arange(0,1,0.001)

message_signal=num.sin(2*num.pi*fm*time)  
carrier_signal=num.sin(2*num.pi*fc*time) 
fm_signal=num.sin(2*num.pi*fc*time+(modulation_index*num.sin(2*num.pi*fm*time)))#Frequency changing w.r.t Message
```

```python
def plotting(time,message_signal,carrier_signal,result_signal,Result_title):
  fig = px.line(x=time, y=message_signal, 
                title='Message Signal',)
  fig.update_layout(
      autosize=False,
      width=800,
      height=400,
      xaxis_title="Time",
      yaxis_title="Amplitude")

  fig.show()

  fig = px.line(x=time, y=carrier_signal,
                title='Carrier Signal')

  fig.update_layout(
      autosize=False,
      width=800,
      height=400,
      xaxis_title="Time",
      yaxis_title="Amplitude")

  fig.update_traces(line_color='#f21d1d')
  fig.show()

  fig = px.line(x=time, y=fm_signal,
                title=Result_title)

  fig.update_layout(
      autosize=False,
      width=800,
      height=400,
      xaxis_title="Time",
      yaxis_title="Amplitude")
  fig.update_traces(line_color='#0c3303')
  fig.show()
```

```python
plotting(time,message_signal,carrier_signal,result_signal=fm_signal,Result_title='FM Signal')
```

<h2 align='center'> AM</h2>


```python
Ac=1
Am=modulation_index*Ac
AM_signal=(1+modulation_index*message_signal)*(Ac*carrier_signal)
```

```python
plotting(time,message_signal,carrier_signal,result_signal=AM_signal,Result_title='AM Signal')
```

<h2 align='center'> PM </h2>

```python
x1=[]
for i in range(len(time)):
    x1.append(1)
x2 = Ac*num.sin(2*num.pi*fm*time)# %generation of sine wave
y = x1*x2; #modulation step
```

```python
plotting(time,message_signal,x2,y,Result_title='PM Signal')
```
